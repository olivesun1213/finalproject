```python
import pandas as pd
```

```python
# Read csv
file_path="Resources/KSI_raw.csv"
ksi=pd.read_csv(file_path)
ksi.head()
```

```python
# Drop columns we do not need
ksi.drop(['YEAR', 'ACCLOC'], axis = 1, inplace=True)
ksi.head()
```

```python
# Drop columns we do not need
ksi.drop(['IMPACTYPE', 'FATAL_NO', 'INITDIR', 'MANOEUVER', 'PEDTYPE', 'PEDACT', 'PEDCOND', 'CYCLISTYPE', 'CYCACT', 'CYCCOND', 'PASSENGER', 'ObjectId'], axis = 1, inplace=True)
ksi.head()
```

```python
# Replace the null with "unknown"
ksi["ROAD_CLASS"].fillna("Unknown", inplace=True)
ksi.loc[ksi["ROAD_CLASS"] == "Unknown"]
```

```python
ksi["LOCCOORD"].fillna("Unknown", inplace=True)
ksi.loc[ksi["LOCCOORD"] == "Unknown"]
```

```python
ksi.fillna("Unknown", inplace=True)
ksi
```

```python
# Check to confirm there is no null values in the dataframe
ksi.isnull().sum()
```

```python
ksi["PEDESTRIAN"]=ksi["PEDESTRIAN"].replace({"Unknown":"No"})
ksi["CYCLIST"]=ksi["CYCLIST"].replace({"Unknown":"No"})
ksi["AUTOMOBILE"]=ksi["AUTOMOBILE"].replace({"Unknown":"No"})
ksi["MOTORCYCLE"]=ksi["MOTORCYCLE"].replace({"Unknown":"No"})
ksi["TRUCK"]=ksi["TRUCK"].replace({"Unknown":"No"})
ksi["TRSN_CITY_VEH"]=ksi["TRSN_CITY_VEH"].replace({"Unknown":"No"})
ksi["EMERG_VEH"]=ksi["EMERG_VEH"].replace({"Unknown":"No"})
ksi["SPEEDING"]=ksi["SPEEDING"].replace({"Unknown":"No"})
ksi["AG_DRIV"]=ksi["AG_DRIV"].replace({"Unknown":"No"})
ksi["REDLIGHT"]=ksi["REDLIGHT"].replace({"Unknown":"No"})
ksi["ALCOHOL"]=ksi["ALCOHOL"].replace({"Unknown":"No"})
ksi["DISABILITY"]=ksi["DISABILITY"].replace({"Unknown":"No"})
ksi
```
